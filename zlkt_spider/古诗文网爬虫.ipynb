## 正则爬取古诗文网

```python
import requests
import re
```

```python
def parse_page(url):
    headers = {
            'User-Agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_12_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/64.0.3282.140 Safari/537.36'
        }
    res = requests.get(url, headers=headers)
    text = res.text
    
    poems = []
    titles = re.findall(r'<div class="yizhu".*?<b>(.*?)</b>', text, re.S)
    dynasties = re.findall(r'<p class="source".*?<a.*?>(.*?)</a>', text, re.S)
    authors =  re.findall(r'<p class="source">.*?<a.*?<a.*?>(.*?)</a>', text, re.S)
    contents = re.findall(r'<div class="contson".*?>(.*?)</div>', text, re.S)
#     print(titles, dynasties, authors)
    contents = [content.strip().replace('<br />', '') for content in contents]
    for i in range(len(titles)):
        poem = {}
        poem['title'] = titles[i]
        poem['dynasty'] = dynasties[i]
        poem['author'] = authors[i]
        poem['content'] = contents[i]
#         print(poem)
        poems.append(poem)
    
    print(poems)
    
        
        


def main():
    for i in range(1,11):
        url = 'http://www.gushiwen.org/default_%s.aspx' % i
        parse_page(url)
    
if __name__ == '__main__':
    main()
```

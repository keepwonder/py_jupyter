## 电影天堂2018新片精品爬取

```python
import requests
from lxml import etree
```

```python
BASE_URL = 'http://dytt8.net/'

def get_html(url): 
    '''
    获取html页面内容
    @params: url 要爬取的页面链接
    @return：返回页面内容
    '''
    headers = {
            'User-Agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_12_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/64.0.3282.140 Safari/537.36'
        }
    res = requests.get(url, headers=headers)
    return res
        

# //div[@class="co_content8"]/ul/table 每一页电影列表
# //div[@class="co_content8"]/ul/table//a/@href 电影详情链接
def get_movie_list(page_num):
    '''
    获取所有电影的链接
    @params: page_num 要爬取的页数
    @return: 返回所有的链接
    '''
    all_movie_link = []
    for i in range(page_num):
        base_url = 'http://dytt8.net/html/gndy/dyzz/list_23_{}.html'.format(i+1)
        res = get_html(base_url)
        html = etree.HTML(res.text)
        page_link_list = html.xpath('//div[@class="co_content8"]/ul//table//a/@href')
        
        # 给每一个url添加主域名前缀
        page_link_list = map(lambda url: BASE_URL+url, page_link_list)
        all_movie_link.extend(page_link_list)
        
#         for movie_link in page_link_list:
#             movie_link = BASE_URL + movie_link
#             all_movie_link.append(movie_link)
            
    return all_movie_link
         
def get_movie_detail(movie_link):
    '''
    获取电影详情页
    @params:movie_link 电影详情页的链接
    '''
    movie = {}
    res = get_html(movie_link)
    html = etree.HTML(res.content)
    # 标题
    title = html.xpath('//font[@color="#07519a"]/text()')[0]
    movie['title'] = title
    movie_html = html.xpath('//div[@id="Zoom"]')[0]
    # 电影海报和截图
#     img = movie_html.xpath('.//img/@src')
#     cover = img[0]
#     screenshot = img[1]

#     movie['cover'] = cover
#     movie['screenshot'] = screenshot
#     电影详情
    movie_detail = movie_html.xpath('.//text()')
    
    def parse_info(info, rule):
        return info.replace(rule, '').strip()
    for index,info in enumerate(movie_detail):
        if info.startswith('◎译　　名'):
            info = parse_info(info,'◎译　　名')
            movie['translate_name'] = info
        elif info.startswith('◎片　　名'):
            info = parse_info(info,'◎片　　名')
            movie['movie_name'] = info
        elif info.startswith('◎年　　代'):
            info = parse_info(info,'◎年　　代')
            movie['year'] = info
        elif info.startswith('◎产　　地'):
            info = parse_info(info,'◎产　　地')
            movie['country'] = info
        elif info.startswith('◎类　　别'):
            info = parse_info(info,'◎类　　别')
            movie['category'] = info
        elif info.startswith('◎语　　言'):
            info = parse_info(info,'◎语　　言')
            movie['language'] = info
        elif info.startswith('◎上映日期'):
            info = parse_info(info,'◎上映日期')
            movie['up_time'] = info
        elif info.startswith('◎豆瓣评分'):
            info = parse_info(info,'◎豆瓣评分')
            movie['douban_score'] = info
        elif info.startswith('◎片　　长'):
            info = parse_info(info,'◎片　　长')
            movie['duration'] = info
        elif info.startswith('◎导　　演'):
            info = parse_info(info,'◎导　　演')
            movie['director'] = info
        elif info.startswith('◎主　　演'):
            info = parse_info(info,'◎主　　演')
            actors = [info]
            for i in range(index+1, len(movie_detail)):
                actor = movie_detail[i].strip()
                if actor.startswith('◎简　　介'):
                    break
                actors.append(actor)
            
            movie['actor'] = actors
        elif info.startswith('◎简　　介'):
            info = parse_info(info,'◎简　　介')
            for i in range(index+1, len(movie_detail)):
                profile = movie_detail[i].strip()
                if profile.startswith('◎'):
                    break
                profile = profile
                movie['profile'] = profile
            
#     下载链接
    try:
        download_url = movie_html.xpath('.//td[@bgcolor="#fdfddf"]/a/@href')[0]
    except:
        download_url = '发生错误'
        print('发生错误')
    
    movie['download_url'] = download_url
    
    print(movie) 
#     return movie

    
if __name__ == '__main__':
#     get_movie_detail('http://dytt8.net//html/gndy/dyzz/20180205/56262.html')

    onepage_list = get_movie_list(1)
    movie_list = []
    for page in onepage_list:
        movie_info = get_movie_detail(page)
        movie_list.append(movie_info)

    print(len(movie_list))
#     for movie in movie_list:
#         print(movie)

```
